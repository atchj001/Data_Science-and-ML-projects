## 1. Data Analysis and Preparation

<i><b>iris</b></i> is one of the mostly used datasets in machine learning. It's a simple and small dataset, with 50 samples of 3 different species of iris. The patel & sepal and <i>lengths</i> & <i>depths</i> (input features) have strong correlations with <i>species</i> (output, responses, labels), easy for data analysis and visualisation, and demonstrating how machine learning models work.

## 2. Machine Learning Models on <i>iris</i>

- The dataset can be easily defined as a classification problem. The data have been pre-processed in sklearn and ready for building different classification models.
- The <i>data</i> part is stored as <i>x</i>, and <i>target</i> part is stored as <i>y</i> (aka: labels, target, response).
- Each row is an observation (aka: sample, example, instance, record). Each column is an input feature (aka: predictor, attribute, independent variable).
- <b>Classification</b> is supervised learning, where the response is discrete / categorical. <b>Regression</b> is also supervised learning, but the response is ordered and continuous.

### A. Decision Tree

### Build and train the decision tree model

```python
# Train a decision tree model
from sklearn.tree import DecisionTreeClassifier

t_model = DecisionTreeClassifier(max_depth = 3)
t_model.fit(x, y) # Training on x (input) against y (output)
```

```python
# Visualise the trained decision tree
from sklearn import tree

plt.figure(figsize=(15,8))
tree.plot_tree(t_model, class_names=iris.target_names, filled=True, fontsize=10) # Visualise the trained decision tree
plt.show()
```

### Evaluate the trained decision tree model

```python
from sklearn import metrics

y_predict = t_model.predict(x)
print(metrics.accuracy_score(y, y_predict))
```

<img align="right" width=280 src="https://upload.wikimedia.org/wikipedia/commons/thumb/1/19/Overfitting.svg/600px-Overfitting.svg.png">

The above accuracy score is very high, meaning that nearly all obervations are correctly calssified by the model! However is this the true measure of the classification accuracy? Is it realistic to expect a model to be correct 100% of the time?<br>

Another potential issue with the training using the whole dataset is that of <b><a href="https://en.wikipedia.org/wiki/Overfitting">overfitting</a></b>. In the figure, a model trained to classify observations with the <font color="green"><b>green curve</b></font> won't perform well on new observations, compared to a model which classifies observations like the <font color="black"><b>black curve</b></font> does.<br>

In overfitting, unwanted noises (rather than the underlying common and general patterns / strutures) in the training data are captured and learned by a model, which won't generalise well to new observations. In machine learning, the <b>k-fold corss-validation</b> method is usually used to avoid overfitting.<br>

### Properly train and test the decision tree model

```python
# Manually partition the data into training and testing datasets, build one model

from sklearn.model_selection import train_test_split

t_model1 = DecisionTreeClassifier(max_depth = 3)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
t_model1.fit(x_train, y_train) # Training on the input of the training partition
print('Training accuracy: \t', metrics.accuracy_score(y_train, t_model1.predict(x_train)))

y_predict = t_model1.predict(x_test) # Testing on the input of the testing partition (new / out-of-sample data)

print('Testing accuracy: \t', metrics.accuracy_score(y_test, y_predict))
```

### Confusing Matrix - Visualising the evaluation on the trained classification model

The function <i>metrics.accuracy_score(y_test, y_predict)</i> calculates the overal accuracy comparing the expected output <i>y_test</i> and <i>y_predict</i>, the output produced by the trained model <i>t_model1</i>. Confusion matrix display the correct and wrong calssifications for different outputs, thus in-sights of how the model performs.

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predict)
cm
```

Question: Does this confusion matrix match that of the above testing accuracy? Is it the same every time? Why?

```python
# Plot the Confusion Matrix
def cf_matrix(mat, t_name):
    
    fig, ax = plt.subplots()
    im = ax.imshow(mat)
    
    ax.set_xticks(range(0,len(t_name)))
    ax.set_yticks(range(0,len(t_name)))
    ax.set_xticklabels(t_name)
    ax.set_yticklabels(t_name)

    for i in range(0,len(t_name)):
        for j in range(0,len(t_name)):
            text = ax.text(j, i, mat[i, j], ha="center", va="center", color="w")
    
    plt.xlabel('true label')
    plt.ylabel('predicted label')
plt.show()
```

```python
cf_matrix(cm, iris.target_names) # Call the self-defined visualisation function
plt.show()
```

### Cross-validation

In <b>k-fold cross-validation</b> (aka rotation estimation or out-of-sample testing), <i>k</i> models are built, each on (<i>k</i>-1) partitions of the data (training data); the remaining one partition is used as testing data. The average accuracy provides a better estimate of the model for the dataset.

```python
from sklearn.model_selection import cross_val_score

t_model2 = DecisionTreeClassifier(max_depth = 5)
print(cross_val_score(t_model2, x, y, cv=10, scoring='accuracy').mean())
```

```python
t_model3 = DecisionTreeClassifier(max_depth = 3)
print(cross_val_score(t_model3, x, y, cv=10, scoring='accuracy').mean())
```

### B. Neural Network

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(5,5), max_iter=3000)
mlp.fit(x_train, y_train)
```

```python
y_predict = mlp.predict(x_test)

print('Testing accuracy: \t', metrics.accuracy_score(y_test, y_predict))
```

```python
x_predict = mlp.predict(x_train)
print('Training accuracy: \t', metrics.accuracy_score(y_train, x_predict))
```

The above accuracy is calculated on one partition of the data (i.e. x_train and x_test). Quite often the accuracy varies between different runs thus may not be accurate evaluating the real performance of a machine learning model.<br>
Cross validation evaluates in one go the accuracy of multiple <i>cv</i> partitions, thus is better evaluating machine learning models.

```python
print(cross_val_score(mlp, x, y, cv=10, scoring='accuracy').mean())
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predict)
cf_matrix(cm, iris.target_names)

plt.show()
```
